```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    tokens = word_tokenize(text)
    tokens = [t for t in tokens if t not in stopwords.words('english')]
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    return tokens
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

def extract_noun_phrases(text):
    doc = nlp(text)
    return [chunk.text for chunk in doc.noun_chunks]
```

```python
import pandas as pd
import spacy

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Sample function
def adjective_detect(text):
    if pd.isna(text):  # handle NaNs
        return ""
    doc = nlp(text)
    adjectives = [token.text for token in doc if token.pos_ == "ADJ"]
    return ' '.join(adjectives)  # Or return adjectives if you want a list
```

```python
df['Q20. “What kind of feeling does this razor pack evoke in you?”'] = df["review"].apply(adjective_detect)
```

```python
df[['Respondent ID','Q20. “What kind of feeling does this razor pack evoke in you?”']].to_csv('q20 adjective packing.csv', index=False)
```
